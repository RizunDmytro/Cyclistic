```python
import pandas as pd
import numpy as np
```

```python
ride = pd.read_csv("2022-divvy-tripdata.csv", delimiter = ",")
ride
#importing data
```

```python
ride.info()
#checking info about data set
```

```python
ride_new = ride.drop(['ride_id','start_station_name','start_station_id','end_station_id','start_lat','start_lng'],axis = 1)
ride_new
#droping tables
```

```python
ride = ride_new.drop(['end_station_name','end_lat','end_lng'], axis = 1)
ride
#dropring more tables
```

```python
ride_new = ride
```

```python
ride_new['started_at'] = pd.to_datetime(ride_new['started_at'])
ride_new['ended_at'] = pd.to_datetime(ride_new['ended_at'])
#changing data type for 2 columns 
```

```python
ride_new
```

```python
ride_new.info()
#checking data type
```

```python
ride_new.count()
#checking missing values
```

```python
ride_time = ride_new['ended_at'] - ride_new['started_at']
ride_time
#calculating ride time
```

```python
def date_to_weekday(date_value):
    return date_value.weekday()
#returning weekday of ride
```

```python
ride_new['started_at'] = ride_new['started_at'].apply(date_to_weekday)
#returning weekday
```

```python
ride_new = pd.concat([ride_new,ride_time], sort=False, axis = 1)
#adding ride_time column to dataset
```

```python
ride_new
#checking dataset
```

```python
ride_new = ride_new.rename(columns = {'rideblePtype':'Bike_type', 'started_at':'Week_day', 'member_casual':'Subscription_type', 0:'Ride_time'})
#renaming columns
```
